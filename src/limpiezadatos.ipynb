```python
#voy a abrir el archivo datos.csv en un dataframe
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
df = pd.read_csv('../data/datos.csv')
```

```python
#voy a borrar las filas que tienen todos los valores nulos
df.dropna(how='all', inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
#creamos una copia del dataframe para trabajar con ella
df2 = df.copy()
```

```python
#en las columnas que empiezan por pg_, voy a copiar los valores del indice 0 en las filas que tienen valores nulos
for col in df2.columns:
    if col.startswith('pg_'):
        df2[col].fillna(df2[col][0], inplace=True)
```

```python
df2.head()
```

```python
#la columna pg_numero_de_publicaciones está en float. Primero la multiplico por 1000 y luego la convierto a int
df2['pg_numero_de_publicaciones'] = df2['pg_numero_de_publicaciones'].apply(lambda x: int(x*1000))

```

```python
df2.head()
```

```python
#la columna pg_seguidores está en string. Voy a convertirlo a int. Si tiene M, lo multiplico por 1000000. Si tiene K, lo multiplico por 1000
def convertir_a_int(x):
    if x.endswith('M'):
        return int(float(x[:-1])*1000000)
    elif x.endswith('K'):
        return int(float(x[:-1])*1000)
    else:
        return int(x)
df2['pg_seguidores'] = df2['pg_seguidores'].apply(convertir_a_int)
```

```python
df2.head()
```

```python
#la columna pg_seguidos está en float. La voy a pasar a int
df2['pg_seguidos'] = df2['pg_seguidos'].apply(lambda x: int(x))
```

```python
df2.head()
```

```python
#elimino la primera fila porque tiene todos los valores nulos
df2.drop(0, inplace=True)
```
