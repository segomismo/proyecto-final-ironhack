```python
#voy a abrir el archivo datos.csv en un dataframe
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
df = pd.read_csv('../data/datos.csv')
```

```python
#la columna pg_numero_de_publicaciones está en float. Compruebo que tenga algun decimal, y si es así, lo multiplico por 1000. Luego lo paso a int
df2['pg_numero_de_publicaciones'] = df2['pg_numero_de_publicaciones'].apply(lambda x: int(x*1000) if x%1 != 0 else int(x))
```

```python
#la columna pg_seguidores está en string. Voy a convertirlo a int. Si tiene M, lo multiplico por 1000000. Si tiene K, lo multiplico por 1000
def convertir_a_int(x):
    if x.endswith('M'):
        return int(float(x[:-1])*1000000)
    elif x.endswith('K'):
        return int(float(x[:-1])*1000)
    else:
        return int(x)
df2['pg_seguidores'] = df2['pg_seguidores'].apply(convertir_a_int)
```

```python
df2.head()
```

```python
#la columna pg_seguidos está en float. La voy a pasar a int
df2['pg_seguidos'] = df2['pg_seguidos'].apply(lambda x: int(x))
```

```python
#quiero unificar todo el codigo que he hecho en una funcion
df_copy = df.copy()

def convertir_a_int(x):
    if x.endswith('M'):
        return int(float(x[:-1])*1000000)
    elif x.endswith('K'):
        return int(float(x[:-1])*1000)
    else:
        return int(x)

def limpieza(df):
    df.dropna(how='all', inplace=True)
    for col in df.columns:
        if col.startswith('pg_'):
          df[col].fillna(df[col][0], inplace=True)
    df['pg_seguidores'] = df['pg_seguidores'].apply(convertir_a_int)
    df['pg_numero_de_publicaciones'] = df2['pg_numero_de_publicaciones'].apply(lambda x: int(x*1000) if x%1 != 0 else int(x))
    df['pg_seguidos'] = df['pg_seguidos'].apply(lambda x: int(x))
    df.drop(0, inplace=True)
    df.reset_index(drop=True, inplace=True)

    df.dropna(subset=['ph_likes'], inplace=True)
    df['ph_likes'] = df['ph_likes'].apply(lambda x: int(x.split(' ')[0].replace('.', '')))
    df['ph_hora'] = df['ph_fecha'].apply(lambda x: x.split('T')[1])
    df['ph_fecha'] = df['ph_fecha'].apply(lambda x: x.split('T')[0])
    df['ph_hora'] = df['ph_hora'].apply(lambda x: x.split(':')[0])
    df['ph_fecha'] = pd.to_datetime(df['ph_fecha'])
    df['ph_hashtags'] = df['ph_descripcion'].apply(lambda x: [i for i in x.split() if i.startswith('#')])
    df['ph_numerado'] = df['ph_numerado'].apply(lambda x: int(x))
    
    df['ph_persona'] = 0
    df['ph_naturaleza'] = 0
    df['ph_animal'] = 0
    df['ph_text'] = 0
    df['ph_R'] = 0
    df['ph_G'] = 0
    df['ph_B'] = 0
    df['ph_WB'] = 0
 
    df['ph_persona'] = df['ph_atributos'].apply(lambda x: 1 if 'people' in x or 'person' in x else 0)
    df['ph_naturaleza'] = df['ph_atributos'].apply(lambda x: 1 if 'nature' in x or 'water' in x or 'trees' in x or 'forest' in x or 'flower' in x else 0)
    df['ph_animal'] = df['ph_atributos'].apply(lambda x: 1 if 'animal' in x or 'cat' in x or 'dog' in x or 'bird' in x or 'fish' in x or 'horse' in x else 0)
    df['ph_text'] = df['ph_atributos'].apply(lambda x: 1 if 'text' in x else 0)
    return df
```

```python
#ahora una funcion que rellene ph_persona, ph_naturaleza, ph_animal, ph_text
def rellenar(df):
    df['ph_persona'] = df['ph_atributos'].apply(lambda x: 1 if 'people' in x or 'person' in x else 0)
    df['ph_naturaleza'] = df['ph_atributos'].apply(lambda x: 1 if 'nature' in x or 'water' in x or 'trees' in x or 'forest' in x or 'flower' in x else 0)
    df['ph_animal'] = df['ph_atributos'].apply(lambda x: 1 if 'animal' in x or 'cat' in x or 'dog' in x or 'bird' in x or 'fish' in x or 'horse' in x else 0)
    df['ph_text'] = df['ph_atributos'].apply(lambda x: 1 if 'text' in x else 0)
    return df
```
