```python
#voy a abrir el archivo datos.csv en un dataframe
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
df = pd.read_csv('../data/datos.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df2 = df.copy()
```

```python
#voy a borrar las filas que tienen todos los valores nulos
df2.dropna(how='all', inplace=True)
```

```python
df2.shape
```

```python
df2.head()
```

```python
#en las columnas que empiezan por pg_, voy a copiar los valores del indice 0 en las filas que tienen valores nulos
for col in df2.columns:
    if col.startswith('pg_'):
        df2[col].fillna(df2[col][0], inplace=True)
```

```python
df2.head()
```

```python
#la columna pg_numero_de_publicaciones está en float. Compruebo que tenga algun decimal, y si es así, lo multiplico por 1000. Luego lo paso a int
df2['pg_numero_de_publicaciones'] = df2['pg_numero_de_publicaciones'].apply(lambda x: int(x*1000) if x%1 != 0 else int(x))
```

```python
df2.head()
```

```python
#la columna pg_seguidores está en string. Voy a convertirlo a int. Si tiene M, lo multiplico por 1000000. Si tiene K, lo multiplico por 1000
def convertir_a_int(x):
    if x.endswith('M'):
        return int(float(x[:-1])*1000000)
    elif x.endswith('K'):
        return int(float(x[:-1])*1000)
    else:
        return int(x)
df2['pg_seguidores'] = df2['pg_seguidores'].apply(convertir_a_int)
```

```python
df2.head()
```

```python
#la columna pg_seguidos está en float. La voy a pasar a int
df2['pg_seguidos'] = df2['pg_seguidos'].apply(lambda x: int(x))
```

```python
df2.head()
```

```python
#elimino la primera fila porque tiene todos los valores nulos
df2.drop(0, inplace=True)
```

```python
df2.head()
```

```python
#reiniciamos el indice
df2.reset_index(drop=True, inplace=True)
```

```python
#vamos a ver si hay algun nulo en la columna ph_likes  
df2['ph_likes'].isnull().sum()
```

```python
#vamos a localizar las filas que tienen nulos en la columna ph_likes
df2[df2['ph_likes'].isnull()]
```

```python
#como el numero de likes es importante, vamos a eliminar las filas que tienen nulos en la columna ph_likes
df2.dropna(subset=['ph_likes'], inplace=True)
```

```python
#ahora vamos a ver si hay algun nulo en la columna ph_likes
df2['ph_likes'].isnull().sum()
```

```python
df2.head()
```

```python
#ahora en la columna ph_likes, vamos a eliminar el punto, la parte que pone "Me gusta" y pasar el valor a int
df2['ph_likes'] = df2['ph_likes'].apply(lambda x: int(x.split(' ')[0].replace('.', '')))
```

```python
df2.head()
```

```python
#vamos a comprobar si hay nulos en ph_descripcion
df2['ph_descripcion'].isnull().sum()
```

```python
#tambien vamos a hacer lo mismo en ph_fecha
df2['ph_fecha'].isnull().sum()
```

```python
#tambien en ph_numerado
df2['ph_numerado'].isnull().sum()
```

```python
#en la columna ph_fecha, que tiene un formato datatime con formato 2023-03-14T20:44:17.000Z, voy a eliminar la parte que pone .000Z
#luego quiero separarlo por la T, y quedarme con la parte de la izquierda, que es la fecha por un lado, y la hora por otro en una columna nueva
df2['ph_hora'] = df2['ph_fecha'].apply(lambda x: x.split('T')[1])
df2['ph_fecha'] = df2['ph_fecha'].apply(lambda x: x.split('T')[0])




```

```python
#ahora en ph_hora, voy a quedarme solo con la hora, sin los minutos ni los segundos
df2['ph_hora'] = df2['ph_hora'].apply(lambda x: x.split(':')[0])
```

```python
#en ph_fecha, voy a convertirlo a formato datetime
df2['ph_fecha'] = pd.to_datetime(df2['ph_fecha'])
```

```python
df2.head()
```

```python
df2.ph_fecha.dtype
```

```python
#en ph_descripcion, voy a coger las palabras que tengan delante un "#" y ponerlas en una nueva columna llamada "ph_hashtags"
df2['ph_hashtags'] = df2['ph_descripcion'].apply(lambda x: [i for i in x.split() if i.startswith('#')])
```

```python
df2.head(20)
```

```python
#la columna ph_numerado quiero convertirlo a int
df2['ph_numerado'] = df2['ph_numerado'].apply(lambda x: int(x))
```
