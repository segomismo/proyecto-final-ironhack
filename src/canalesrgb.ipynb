```python
from PIL import Image
import numpy as np
import pandas as pd

df = pd.read_csv('../data/datos1_clean.csv')
```

```python
df_copy = df.copy()
```

```python
for i in df.ph_nombre:
    imagen = Image.open('../images/'+i)

    try:
        r, g, b = imagen.split()
        r_array = np.array(r)
        g_array = np.array(g)
        b_array = np.array(b)

        media_r = np.mean(r_array)
        media_g = np.mean(g_array)
        media_b = np.mean(b_array)
        if media_r > media_g and media_r > media_b:
            df.loc[df.ph_nombre == i, 'ph_R'] = 1
        elif media_g > media_r and media_g > media_b:
            df.loc[df.ph_nombre == i, 'ph_G'] = 1
        elif media_b > media_r and media_b > media_g:
            df.loc[df.ph_nombre == i, 'ph_B'] = 1
    except:
         df.loc[df.ph_nombre == i, 'ph_WB'] = 1

```
